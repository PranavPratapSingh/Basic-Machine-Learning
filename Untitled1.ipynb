```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
import cv2
```

```python
im =cv2.imread('img.jpg')
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
r,c=im.shape[:2]
out_r=64
im=cv2.resize(im,(int(out_r*float(c)/r),out_r))
pixels=im.reshape((-1,3))
print pixels.shape
plt.imshow(im)
```

```python
km =KMeans().fit(pixels)
```

```python
colors=np.asarray(km.cluster_centers_,dtype='uint8')
```

```python
plt.figure(0)
for ix in range(colors.shape[0]):
    patch=np.ones((20,20,3))
    patch[:,:,:]=255-colors[ix]
    plt.subplot(1,colors.shape[0],ix+1)
    plt.axis("off")
    plt.imshow(patch)
```
