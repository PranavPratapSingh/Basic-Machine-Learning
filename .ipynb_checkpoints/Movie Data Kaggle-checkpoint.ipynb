```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
ds=pd.read_csv('../Datasets//movie_metadata.csv')
ds.describe()
```

```python
print ds.columns
```

```python
df=ds
```

```python
df.head()
```

```python
cols_to_drop=[
    'color',
    'movie_imdb_link',
    'Unnamed: 28',
]
df=df.drop(cols_to_drop,axis=1)
```

```python
plt.figure(0)
sns.heatmap(df.corr())
plt.show()
```

```python
label = df['imdb_score'].values
data = df[df.columns[1:]].values
```

```python
le=LabelEncoder()
dq=df[df.columns].apply(le.fit_transform)
data=dq[dq.columns[1:]].values
```

```python
label=dq[dq.columns[0]].values
```

```python
split=int(0.8*data.shape[0])
X_train=data[:split]
X_test=data[split:]

y_train=label[:split]
y_test=label[split:]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DT =DecisionTreeRegressor().fit(X_train,y_train)
```

```python
DT.score(X_test,y_test)
```

```python
dict={}
for ix in range(len(df.columns)-1):
    if(DT.feature_importances_[ix] < 0.01):
        dict[df.columns[ix]]= DT.feature_importances_[ix]
print dict
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor(n_estimators=50).fit(X_train,y_train)
print RF.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression().fit(X_train,y_train)
```

```python
dist=[]
for ix in range(len(y_test)):
    if abs(lr.predict(X_test[ix].reshape(1,-1)) - y_test[ix])> 5:
        dist.append([ix , lr.predict(X_test[ix]) - y_test[ix]])
```

```python
print float(len(dist))/len(y_test)
```
